```python
# Group by brand and calculate average MSRP
avg_price_per_brand = df.groupby('Make')['MSRP'].mean().round()
```

```python
top_10_brands
```

```python
# Plot
plt.figure(figsize=(10,6))
top_10_brands.plot(kind='barh', color='skyblue')
plt.xlabel('Average MSRP ($)')
plt.title('Top 10 Car Brands by Average MSRP')
plt.gca().invert_yaxis()  # Highest at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Group by Year and calculate average MSRP
avg_price_per_year = df.groupby('Year')['MSRP'].mean().round()

# Sort by year (just in case)
avg_price_per_year = avg_price_per_year.sort_index()
```

```python
avg_price_per_year
```

```python
# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(avg_price_per_year.index, avg_price_per_year.values, marker='o', color='green')
plt.title('Average MSRP Over the Years')
plt.xlabel('Year')
plt.ylabel('Average MSRP ($)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Clean column names (remove leading/trailing spaces if any)
df.columns = df.columns.str.strip()

# Convert columns to numeric if needed
df['City MPG'] = pd.to_numeric(df['city mpg'], errors='coerce')
df['Highway MPG'] = pd.to_numeric(df['highway MPG'], errors='coerce')
df['Engine HP'] = pd.to_numeric(df['Engine HP'], errors='coerce')
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')

# Create Average MPG column
df['Average MPG'] = (df['City MPG'] + df['Highway MPG']) / 2

# Group by Year and calculate average Engine HP and Average MPG
avg_per_year = df.groupby('Year')[['Engine HP', 'Average MPG']].mean().round()

# Sort by Year for proper plotting
avg_per_year = avg_per_year.sort_index()

# Plot both trends
plt.figure(figsize=(12,6))

plt.plot(avg_per_year.index, avg_per_year['Engine HP'], label='Average Engine HP', color='crimson', marker='o')
plt.plot(avg_per_year.index, avg_per_year['Average MPG'], label='Average MPG', color='seagreen', marker='s')

# Chart formatting
plt.title('Engine HP and Fuel Efficiency Over the Years', fontsize=14)
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Filter only for Compact, Midsize, or Large vehicles
size_counts = df['Vehicle Size'].value_counts(normalize=True) * 100

# Filter relevant sizes
target_sizes = size_counts.loc[['Compact', 'Midsize', 'Large']]

# Display percentage
target_sizes.round(2)
```

```python
# Plot pie chart
plt.figure(figsize=(7,7))
plt.pie(target_sizes, labels=target_sizes.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'salmon'])
plt.title('Percentage of Car Sizes: Compact, Midsize, Large')
plt.axis('equal')  # Keeps pie circular
plt.show()
```

```python
# Count occurrences of each vehicle style
style_counts = df['Vehicle Style'].value_counts()

# Display top styles
style_counts
```

```python
# Plot
plt.figure(figsize=(12,6))
style_counts.plot(kind='bar', color='mediumslateblue')

# Formatting
plt.title('Most Sold Vehicle Styles', fontsize=14)
plt.xlabel('Vehicle Style')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Filter cars with MSRP >= 40000 and Popularity >= 1000
filtered_df = df[(df['MSRP'] >= 40000) & (df['Popularity'] >= 1000)]
```

```python
# Count how many such vehicles per brand
brand_counts = filtered_df['Make'].value_counts()

# Display result
brand_counts
```

```python
# Plot
plt.figure(figsize=(12,6))
brand_counts.plot(kind='bar', color='darkcyan')

# Formatting
plt.title('Car Brands with Vehicles ≥ $40,000 and Popularity ≥ 1000')
plt.xlabel('Car Brand')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
